##Base credit data

```python
base_credit_data = Orange.data.Table('/content/drive/MyDrive/Curso Machine Learning /Bases de dados/credit_data_regras.csv')
base_credit_data
```

```python
base_credit_data.domain
```

```python
base_dividida = Orange.evaluation.testing.sample(base_credit_data, n = 0.25)
base_dividida
```

```python
base_credit_treinamento = base_dividida[1]
base_credit_teste = base_dividida[0]
base_credit_treinamento
```

```python
len(base_credit_treinamento), len(base_credit_teste)
```

```python
cn2 = Orange.classification.rules.CN2SDLearner()
regras_credit_data = cn2(base_credit_treinamento)
```

```python
for regras in regras_credit_data.rule_list:
  print(regras)
```

```python
# História boa, dívida alta, garantias nenhuma, renda > 35
# História ruim, dívida alta, garantia adequada, renda < 15

previsoes = Orange.evaluation.testing.TestOnTestData(base_credit_treinamento, base_credit_teste, [lambda testdata: regras_credit_data])
```

```python
previsoes
```

```python
Orange.evaluation.CA(previsoes)
```

Base Credit Data

```python
base_credit = Orange.data.Table('/content/drive/MyDrive/Curso Machine Learning /Bases de dados/credit_data_regras.csv')
```

```python
base_credit.domain
```

```python
majority = Orange.classification.MajorityLearner()
```

```python
previsoes = Orange.evaluation.testing.TestOnTestData(base_credit, base_credit, [majority])
```

```python
Orange.evaluation.CA(previsoes)
```

```python
for registro in base_credit:
  print(registro.get_class())
```

```python
from collections import Counter
Counter(str(registro.get_class()) for registro in base_credit)
```
