##Base de Risco de Crédito

```python
import pickle
with open ('/content/drive/MyDrive/Curso Machine Learning /Bases de dados/risco_credito.pkl', 'rb') as f:
  X_risco_credito, Y_risco_credito = pickle.load(f)
```

##Base de dados Crédito

```python
with open ('/content/drive/MyDrive/Curso Machine Learning /Bases de dados/credit.pkl', 'rb') as f:
  X_credito_treinamento, Y_credito_treinamento, X_credito_teste, Y_credito_teste  = pickle.load(f)
```

```python
X_credito_treinamento.shape, Y_credito_treinamento.shape
```

```python
arvore_credito = DecisionTreeClassifier(criterion='entropy', random_state=0)
arvore_credito.fit(X_credito_treinamento, Y_credito_treinamento)
```

```python
previsoes = arvore_credito.predict(X_credito_teste)
previsoes
```

```python
Y_credito_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(Y_credito_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(arvore_credito)
cm.fit(X_credito_treinamento, Y_credito_treinamento)
cm.score(X_credito_teste,Y_credito_teste)
```

```python
print(classification_report(Y_credito_teste,previsoes))
```

```python
from sklearn import tree
previsores = ['income', 'age', 'loan']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (20,20))
tree.plot_tree(arvore_credito, feature_names=previsores, class_names=str(arvore_credito.classes_),filled=True);
fig.savefig('arvore_credito.png')
```

# Random Forest

```python
from sklearn.ensemble import  RandomForestClassifier
import pickle
```

#Base credit data

```python
with open ('/content/drive/MyDrive/Curso Machine Learning /Bases de dados/credit.pkl', 'rb') as f:
  X_credito_treinamento, Y_credito_treinamento, X_credito_teste, Y_credito_teste  = pickle.load(f)
```

```python
random_forest_credit = RandomForestClassifier(n_estimators=40,criterion='entropy', random_state=0)
random_forest_credit.fit(X_credito_treinamento, Y_credito_treinamento)
```

```python
previsoes = random_forest_credit.predict(X_credito_teste)
previsoes
```

```python
Y_credito_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(Y_credito_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(random_forest_credit)
cm.fit(X_credito_treinamento, Y_credito_treinamento)
cm.score(X_credito_teste,Y_credito_teste)
```

```python
print(classification_report(Y_credito_teste, previsoes))
```
